```python
import pandas as pd
import numpy as np

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('./ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('./ml-1m/ratings.dat', sep='::', header=None, names=rnames,  engine='python')
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('./ml-1m/movies.dat', sep='::', header=None, names=mnames,  engine='python')
data = pd.merge(pd.merge(ratings, users), movies)
```

1- Filter  films that have received at least 250 ratings. **Use only these films in the rest of exercices**

```python
def ex1():
    ratings_by_title = data.groupby('title').count()
    films = ratings_by_title[ratings_by_title >= 250]
    return films.sort_values(by='title',ascending=False)
```

```python
ex1()
```

2- Obtain the mean ratings for each movie grouped by gender that have at least 250 ratings. 

```python
def ex2():
    mean_ratings = data.pivot_table('rating', index='title',columns='gender', aggfunc='mean')
    ratings_by_title = data.groupby('title').size()
    ratings_by_title = ratings_by_title.index[ratings_by_title >= 250]
    mean_ratings = mean_ratings.loc[ratings_by_title]
    return mean_ratings
```

```python
ex2()
```

3- Show films more valued by women.

```python
def ex3():
    mean_ratings = ex2()
    top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
    return top_female_ratings.F
```

```python
ex3()
```

4- Now we wonder which movies are rated more differently between men and women. Which films have more different rating and are more highly valued by women? 

```python
def ex4():
    mean_ratings = ex2()
    mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
    sorted_by_diff = mean_ratings.sort_values(by='diff')
    return sorted_by_diff
```

5- And which films have more different rating and are more highly valued by men? 

```python
def ex5():
    data = ex4()
    return data[::-1]
```

6- What are the films that have generated the most discordant ratings, regardless of gender?

```python
def ex6():
    active_titles = ex1().index
    rating_std_by_title = data.groupby('title')['rating'].std()
    rating_std_by_title = rating_std_by_title.loc[active_titles]
    std = rating_std_by_title.sort_values(ascending=False)
    return std
```

10- Read data from csv file: `ma-ba.csv`. Count the number of times `Barça` wins `Madrid` and compute the stadistics of % win, % lose and % draw.

```python
def ex10():
    maba = pd.read_csv('ma-ba.csv', names=['No','Fecha','Lugar','Partidos','resultado','Competicion'], encoding= 'unicode_escape', header=0)
    df = pd.DataFrame()
    df["Local"]= maba["Partidos"].str.split('-').apply(lambda x: x[0].strip())
    df["Visitor"]= maba["Partidos"].str.split('-').apply(lambda x: x[1].strip())
    df["GL"]= maba["resultado"].str.split('-').apply(lambda x: x[0])
    df["GV"]= maba["resultado"].str.split('-').apply(lambda x: x[1])

    df["Local"].replace(to_replace=["Madrid FC","R.Madrid"], value = ["R. Madrid"]*2, inplace=True)
    df["Visitor"].replace(to_replace= ["Madrid FC","R.Madrid"], value = ["R. Madrid"]*2, inplace=True)

    wins_barcelona = df[((df['Local'] == 'Barcelona') & (df['GL'] > df['GV'])) | (df['Visitor'] == 'Barcelona') & (df['GV'] > df['GL'])]
    lose_barcelona = df[((df['Local'] == 'Barcelona') & (df['GL'] < df['GV'])) | (df['Visitor'] == 'Barcelona') & (df['GV'] < df['GL'])]

    wins_madrid = df[((df['Local'] == 'R. Madrid') & (df['GL'] > df['GV'])) | (df['Visitor'] == 'R. Madrid') & (df['GV'] > df['GL'])]
    lose_madrid = df[((df['Local'] == 'R. Madrid') & (df['GL'] < df['GV'])) | (df['Visitor'] == 'R. Madrid') & (df['GV'] < df['GL'])]


    draw = df[(df['GL'] == df['GV'])]

    wins_barcelona_percentage = (len(wins_barcelona) / len(df))
    lose_barcelona_percentage = (len(lose_barcelona) / len(df))
    wins_madrid_percentage = (len(wins_madrid) / len(df))
    lose_madrid_percentage = (len(lose_madrid) / len(df))

    draw_percentage = (len(draw) / len(df))

    data = {'Name':['Barcelona', 'R. Madrid'],
          'Wins':[len(wins_barcelona), len(wins_madrid)], 
          'lose':[len(lose_barcelona), len(lose_madrid)],
          'draw':[len(draw), len(draw)],
          'win_per':[wins_barcelona_percentage, wins_madrid_percentage],
          'lose_per':[lose_barcelona_percentage, lose_madrid_percentage],
          'draw_per':[draw_percentage, draw_percentage]}

    result = pd.DataFrame(data)
    return result



```
